Axes, Labels, Legend, Grid


```python
import matplotlib.pyplot as plt
%matplotlib inline
days = [1,2,3,4,5,6,7]
max_t =[50, 51, 52,48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t  = [45, 48 , 48 , 46, 42, 41,42]
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")
plt.plot(days, max_t, label= 'max')
plt.plot(days, min_t, label= 'min')
plt.plot(days, avg_t, label= 'avg')
# Legend property sets the label.
# Using loc in legend function we can change the position of the legend.
# By default legend is moved to the blank area.
plt.legend(loc= "upper right")
# Create grid using grid property.
plt.grid()
```

Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
company = ['Google', 'Amazon', 'MSFT', 'FB']
revenue = [90, 136 ,89, 27]
profit = [40, 2, 34, 12]
# Creating the array of range of list.
ypos = np.arange(len(company))
# Using xticks 
plt.xticks(ypos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos, revenue, label ="Revenue")
plt.bar(ypos, profit, label = "Profit")
plt.legend()
```

```python

# horizontal bar graph.
plt.xticks(ypos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.barh(ypos, revenue, label ="Revenue")
plt.barh(ypos, profit, label = "Profit")
plt.legend()
```

Histogram

```python
# Histograms are barcharts where the y-axis contains the frequency of a variable.
bloodsugar_men= [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
bloodsugar_women = [113, 80, 50, 220, 49, 88, 93, 115, 45, 80, 27, 89, 129]
# By default there are 10 bars/bins.
plt.hist([bloodsugar_men, bloodsugar_women], bins = 3, rwidth = 0.95, histtype='step', color = ['green', 'orange'],
         label = ['men', 'women']
         )
plt.legend()
# Histtype could be set to step so it looks like a line.
```

Pie Chart

```python
exp_values = [1400, 600, 300, 410, 250, 359]
exp_labels = ["home", 'rent', 'food', "phone/internet bill", "car", "other utilities"]
# Autopct is used define the type of decimal used. 
# Using explode property we can move the required pie out of the graph.
# Using startangle plroperty we can rotate the angle of the pie chart counterclockwise.
plt.pie(exp_values, labels = exp_labels, autopct= "%0.2f%%", shadow = True, explode = [0,0.2,0,0,0, 0], startangle = 180)
```
